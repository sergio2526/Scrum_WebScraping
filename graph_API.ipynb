```

def extract_data_posts(token, graph):

  fields = ['posts']
  fields =','.join(fields)
  data = graph.get_object(id_cuenta, fields=fields)

  with open('data.json', 'w') as file:
    return json.dump(data, file, indent=4)

extract_data_posts(token, graph)

```

### Normalizando Data

```
def normalize_data_posts():

  path = r'data.json'

  with open(path) as data:    
      d = json.load(data)  
      
  df_message_posts = pd.json_normalize(d['posts']['data'])
  return df_message_posts

normalize_data_posts()
```

### Data Comments

```
def extract_data_comments(token, graph):

  fields = ['posts{comments}']

  fields =','.join(fields)

  data = graph.get_object(id_cuenta, fields=fields)

  with open('data_comments.json', 'w') as file:
    return json.dump(data, file, indent=4)


extract_data_comments(token, graph)
```

```

def normalize_data_posts_comments():

  path = r'data_comments.json'
  with open(path) as data:    
      d = json.load(data)    
      
  df_comments = pd.json_normalize(d['posts']['data'])
  df_comments.rename(columns={'comments.data':'comments'},inplace=True)
  return df_comments

normalize_data_posts_comments()
```

```
df_comments = normalize_data_posts_comments()

def normalize_data_comments():
  created_time_comentario, message_comentario, id_comentario, lista_commentarios = [],[],[],[]

  #Guardando en una lista y aplicando la Transpuesta
  #esto se hace para darle un mejor orden a los datos de la columna comments

  for i in range(22):
    lista = df_comments.comments.T[i]
    lista_commentarios.append(lista)

  # Guardando valores en las respectivas listas
  for i in range(len(lista_commentarios)):
    if str(lista_commentarios[i]) != 'nan':
      for elem in lista_commentarios[i]:    
        for k,v in elem.items():       
            if k == 'message':
              message_comentario.append(v)
            if k == 'id':
              id_comentario.append(v)
            if k == 'created_time':
              created_time_comentario.append(v)


  # generando un DataFrame con los datos 
  df_comments_post = pd.DataFrame({"created_time_comentario":created_time_comentario,"Id_comentario":id_comentario, "Comentarios":message_comentario})
  df_comments_post.tail(10)

  return df_comments_post

normalize_data_comments()
```
